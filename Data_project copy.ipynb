```python
import matplotlib
import matplotlib.pyplot as plt
gas = pd.read_csv('airlineflights.csv', parse_dates=["Year"])
#plt.plot(gas.Year, gas.Max_Seats)
gas.groupby("Year").sum().All_Flights.plot()
plt.figure(figsize=(8,5))

plt.show()
```

```python
for name, df in gas.groupby("Australian_City") :
    df.groupby("Year").sum().All_Flights.plot(label=name)
plt.legend(loc="upper center", bbox_to_anchor=(1.25, 1), ncol=2)
```

```python
import matplotlib
import matplotlib.pyplot as plt
gas = pd.read_csv('airlineflights.csv', parse_dates=["Year"])
#plt.plot(gas.Year, gas.Max_Seats)
gas.groupby("Year").sum().Max_Seats.plot()
plt.figure(figsize=(8,5))

plt.show()
```

```python
for name, df in gas.groupby("Australian_City") :
    df.groupby("Year").sum().Max_Seats.plot(label=name)
plt.legend(loc="upper center", bbox_to_anchor=(0.5, 1.15), ncol=2)
```

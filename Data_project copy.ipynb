```python
import matplotlib
import matplotlib.pyplot as plt
gas = pd.read_csv('AirFlights.csv')
plt.plot(gas.Year, gas.TATOTAL)

plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
gas = pd.read_csv('AirFlights.csv', parse_dates=["Year"])
#plt.plot(gas.Year, gas.Max_Seats)
gas.groupby("Year").sum().TATOTAL.plot()
plt.figure(figsize=(8,5))

plt.show()
```

```python
for name, df in gas.groupby("AIRPORT") :
    df.groupby("Year").sum().TATOTAL.plot(label=name)
plt.legend(loc="upper center", bbox_to_anchor=(1.25, 1), ncol=2)
```

```python
import matplotlib
import matplotlib.pyplot as plt
gas = pd.read_csv('AirFlights.csv', parse_dates=["Year"])
#plt.plot(gas.Year, gas.Max_Seats)
gas.groupby("Year").sum().TATOTAL.plot()
plt.figure(figsize=(8,5))

plt.show()
```

```python
for name, df in gas.groupby("Australian_City") :
    df.groupby("Year").sum().Max_Seats.plot(label=name)
plt.legend(loc="upper center", bbox_to_anchor=(0.5, 1.15), ncol=2)
```
